Splitting and labelling genres

```python
a=genre.unique()
b=[]
for i in a:
    c=i.split('|')
    for j in c:
        b.append(j)
```

```python
gen = list(set(b))
gen_lab = range(len(gen))
Gen = {}
for i in range(len(gen)):
    Gen.update({gen[i]:gen_lab[i]})
Gen
```

```python
direc = data.pop('director_name').reset_index(drop=True)
```

```python
direc
```

Data for Genre prediction

```python
GENRE = []
DIR = []
for i in range(len(genre)):
    g = genre[i].split('|')
    for j in g:
        DIR.append(direc[i])
        GENRE.append(Gen[j])
DIR = np.array(DIR)
GENRE = np.array(GENRE)
```

```python
l = int(0.8*len(GENRE))
GenTrain = GENRE[:l]
GenTest = GENRE[l:]
DirTrain = DIR[:l]
DirTest = DIR[l:]
```

Model

```python
GenPred = []
K = 12
for i in DirTest:
    #d = np.argmin(abs(DirTrain-i))
    d = abs(DirTrain-i)
    tup = []
    for j in range(len(DirTrain)):
        tup.append((d[j],j))
    g = sorted(tup,key=lambda x:x[0])[:K]
    gr = []
    for m,j in g:
        gr.append(GenTrain[j])
    v,_ = np.unique(gr,return_counts=True)
    GenPred.append(v[np.argmax(_)])
```

```python
#loss=sum((GenPred-GenTest)**2)/len(GenTest)
acc=0
for i in range(len(GenPred)):
    if GenPred[i]==GenTest[i]:
        acc+=1
acc/len(GenPred)
```
