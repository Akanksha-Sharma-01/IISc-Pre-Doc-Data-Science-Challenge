```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
import numpy as np
from sklearn.model_selection import train_test_split
```

Reading Data

```python
data = pd.read_csv("D:\Akanksha\Prep\PhD\Pre-Doc IISc\p1_movie_metadata.csv")
```

Removing rows with NaN values

```python
data.dropna(inplace=True)
data.reset_index(drop=True)
```

Correlation among features

```python
CORR = data.corr()
CORR
```

Labeling directors name

```python
dir = data.director_name.unique()
dir_lab = range(np.shape(dir)[0])
for i in range(np.shape(dir)[0]):
    data.loc[data['director_name']==dir[i],'director_name']=dir_lab[i]
#data['director_name'] = data['director_name'].astype(float)
```

Choosing features for year prediction

```python
XT = data[['director_name','num_critic_for_reviews','actor_3_facebook_likes','cast_total_facebook_likes','actor_2_facebook_likes','aspect_ratio','movie_facebook_likes']]
YT = data['title_year']
```

```python
XT.columns = XT.columns.astype(str)
```

Calculating mean of all features

```python
MU = XT.groupby(['director_name']).mean()
MU
```

Splitting

```python
XT_train, XT_test, YT_train, YT_test = train_test_split(XT, YT, test_size=0.2)
```

```python
XT_test = XT_test.reset_index(drop=True)
YT_test = YT_test.reset_index(drop=True)
```

Model

```python
model = RandomForestRegressor(n_estimators=300, random_state=42)
model.fit(XT_train, YT_train)
```

Testing

```python
XT_test = XT_test['director_name']
```

```python
X = []
for i in range(len(XT_test)):
    X.append(pd.concat([pd.Series(data=XT_test.iloc[i]),MU.iloc[XT_test.iloc[i]]],axis=0))
X = np.array(X)
```

```python
X
```

```python
PredYear = np.int_(model.predict(X))
YearLoss = sum(((PredYear-YT_test)**2)/len(YT_test))
```
